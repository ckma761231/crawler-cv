```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text , "html5lib")
posts = soup.find_all('a', class_='PostEntry_root_V6g0rd')
count = 0
for soup1 in posts:
    title = soup1.find('h3', class_='Title__Text-v196i6-0 gmfDU')
    if title:
        count += 1
        print(count)
        print(title.text)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text)

posts = soup.find_all(class_='PostEntry_root_V6g0rd')
for post in posts:
    post_title = post.find_all(class_='PostEntry_excerpt_2eHlNn')
    if post_title:
        print(post_title)
```

### 3. 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？

```python
import requests
url = 'https://www.zhihu.com/explore'
headers = {'user-agent': 'my-app/0.0.1'}
'''
Your Code Here
'''
r = requests.get(url,headers =headers)

r.encoding = 'utf-8'
print(r.text)
```

```python
soup = BeautifulSoup(r.text)
posts = soup.find_all('a', class_='ExploreSpecialCard-title')
for p in posts:
    print(p.text)
```
