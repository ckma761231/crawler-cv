### ① 取出今天所有的發文

```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text)
```

```python
box = {'date':[],'cls':[],'title':[]}
for i in soup.find(class_="part_list_2").find_all('h3'):
    date = i.find(class_="date").text
    cls  = i.find('em').text
    title= i.find('a').text
    box['date'].append([date])
    box['cls'].append([cls])
    box['title'].append([title])
```

```python
box2 = []
for i in soup.find(class_="part_list_2").find_all('h3'):
    date = i.find(class_="date").text
    cls  = i.find('em').text
    title= i.find('a').text
    box2.append([date, cls, title])
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
Your Code
'''
import pandas as pd

table = pd.DataFrame(box, columns=['date','cls','title'])
table
```
